```python
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx
```

```python
#print all graph info
def displayGraphInfo(graph):
    print('Nodes:',graph.node)
    print('Edges:',graph.edges)
    print('Degrees:',graph.degree)
    print('Neighborhoods:',[{i: [*graph.neighbors(i)]} for i in range(len(graph.node))])
```

```python
# simple graph
s = nx.Graph()

s.add_edges_from([[0,i] for i in range(1,4)])
s.add_cycle([1,2])

nx.draw(s, with_labels=True, font_size=20, font_color='white')

displayGraphInfo(s)
```

```python
#complete graph
k = nx.complete_graph(4)

k.add_edges_from([[0,i] for i in range(1,4)])
k.add_cycle([1,2])

nx.draw(k, with_labels=True, font_size=20, font_color='white')

displayGraphInfo(k)
```

```python
#cycle graph
c = nx.cycle_graph(6)


nx.draw(c, with_labels=True, font_size=20, font_color='white')

displayGraphInfo(c)
```

```python
#wheel graph
w = nx.wheel_graph(6)


nx.draw(w, with_labels=True, font_size=20, font_color='white')

displayGraphInfo(w)
```

```python
#adjacency matrix and bipartite testing

mat = np.matrix('0 1 1 1; 1 0 1 1; 1 1 0 1; 1 1 1 0')
a = nx.from_numpy_matrix(mat)

nx.draw(a, with_labels=True, font_size=20, font_color='white')

print(nx.is_bipartite(a))

displayGraphInfo(a)
```
